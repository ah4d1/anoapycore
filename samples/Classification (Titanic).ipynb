#### Train-Test Split

```python
# set feature columns
```

```python
loc_x = ap.data.column.delete(v_data_training,'Survived')
```

```python
ap.data.sample(loc_x)
```

```python
# set target column
```

```python
loc_y = ap.data.select(v_data_training,'Survived')
```

```python
ap.data.sample(loc_y)
```

```python
# split training dataset into train & test
```

```python
v_splitted_data = ap.mlearn.data.split(loc_x,loc_y)
```

```python
ap.data.stat.desc(v_splitted_data.x_train)
```

```python
ap.data.stat.desc(v_splitted_data.x_test)
```

### Logistic Regression

```python
# train model from training dataset (train & test)
```

```python
v_log_reg = ap.mlearn.classification.logreg.run(v_splitted_data.x_train,v_splitted_data.x_test, \
                                                         v_splitted_data.y_train,v_splitted_data.y_test)
```

```python
v_log_reg.model
```

```python
# performance of the model
print('Train')
print(v_log_reg.evals['train'].all)
print('Test')
print(v_log_reg.evals['test'].all)
```

```python
# performance of the model
print('Train')
print('Accuracy : ' + str(v_log_reg.evals['train'].accuracy))
print('Precision : ' + str(v_log_reg.evals['train'].precision))
print('Recall : ' + str(v_log_reg.evals['train'].recall))
print('Test')
print('Accuracy : ' + str(v_log_reg.evals['test'].accuracy))
print('Precision : ' + str(v_log_reg.evals['test'].precision))
print('Recall : ' + str(v_log_reg.evals['test'].recall))
```

```python
print(v_log_reg.report)
```

```python
v_log_reg.roc_chart
```

```python
# Predict
```

```python
loc_predict_logreg = ap.mlearn.classification.logreg.predict(v_log_reg,v_data_testing)
```

```python
loc_predict_logreg.prediction
```

```python
loc_predict_logreg.worksheet
```

```python
ap.data.save.xls(a_data=loc_predict_logreg.worksheet,a_filename='D:/_result/classification_logreg_titanic.xlsx')
```

### Support Vector Classifier
