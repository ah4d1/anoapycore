```python
# original file : https://www.kaggle.com/alphaepsilon/housing-prices-dataset
v_data_train = ap.data.load.csv("D:/_data/house_price/train.csv",",")
v_data_test = ap.data.load.csv("D:/_data/house_price/test.csv",",")
```

### Linear Regression

```python
v_feature_1 = ['LotArea','YrSold']
v_feature_2 = ['LotArea','YrSold','LotFrontage']
v_target = 'SalePrice'
```

```python
# Sample 1
```

```python
v_model_1 = ap.mlearn.regression.linreg.run(a_data=v_data_train,a_feature=v_feature_1,a_target=v_target)
```

```python
v_model_1.model
```

```python
v_model_1.coefficient
```

```python
v_model_1.intercept
```

```python
v_model_1.r_square
```

```python
v_model_1.rmse
```

```python
v_model_2 = ap.mlearn.regression.linreg.run(a_data=v_data_train,a_feature=v_feature_2,a_target=v_target)
```

```python
print(v_model_2.model)
print(v_model_2.coefficient)
print(v_model_2.intercept)
print(v_model_2.r_square)
print(v_model_2.rmse)
```

### Predict Response

```python
ap.data.dimension(v_data_test)
```

```python
ap.data.sample(v_data_test)
```

#### Use Model 1

```python
y_pred_1 = ap.mlearn.regression.linreg.predict(a_model=v_model_1,a_data=v_data_test,a_features=v_feature_1)
```

```python
y_pred_1.prediction
```
